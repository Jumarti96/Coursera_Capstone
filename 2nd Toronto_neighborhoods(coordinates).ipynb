```python
# We load the csv file with the coordinates.
url = "https://cocl.us/Geospatial_data"
coordinates = pd.read_csv(url)
coordinates.head(10)
```

```python
# Both dataframes are sorted differently, so we perform a code to gather the coordinates and add them to the lists created previously in the order of the main dataframe.
for index, row in toronto_neighborhoods.iterrows():
    if row['Postal Code'] in coordinates['Postal Code'].tolist():
        i = coordinates['Postal Code'].tolist().index(row['Postal Code'])
        Latitude.append(coordinates['Latitude'][i])
        Longitude.append(coordinates['Longitude'][i])
```

```python
# We add the lists with our coordinates to the main dataframe.
toronto_neighborhoods['Latitude'] = Latitude
toronto_neighborhoods['Longitude'] = Longitude
toronto_neighborhoods.head(10)
```
